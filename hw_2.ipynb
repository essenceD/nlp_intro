1. Создайте мешок слов с помощью sklearn.feature_extraction.text.CountVectorizer.fit_transform(). Применим его к 'tweet_stemmed' и 'tweet_lemmatized' отдельно.</br>
●	Игнорируем слова, частота которых в документе строго превышает порог 0.9 с помощью max_df.</br>
●	Ограничим количество слов, попадающий в мешок, с помощью max_features = 1000.</br>
●	Исключим стоп-слова с помощью stop_words='english'.</br>
●	Отобразим Bag-of-Words модель как DataFrame. columns необходимо извлечь с помощью CountVectorizer.get_feature_names().

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_pickle('preprocessed_tweets.pkl')
df
```

```python
df['tweet_stemmed'] = df.tweet_stemmed.apply(lambda x: ' '.join(word for word in x))
```

```python
vectorizer_stemmed = CountVectorizer(stop_words='english', max_df=0.9, max_features=1000)
bag_of_words_stemmed = vectorizer_stemmed.fit_transform(df['tweet_stemmed'])
```

```python
feature_names_stemmed = vectorizer_stemmed.get_feature_names()
pd.DataFrame(bag_of_words_stemmed.toarray(), columns = feature_names_stemmed)
```

```python
df['tweet_lemmatized'] = df.tweet_lemmatized.apply(lambda x: ' '.join(word for word in x))

vectorizer_lemmatized = CountVectorizer(stop_words='english', max_df=0.9, max_features=1000)
bag_of_words_lemmatized = vectorizer_lemmatized.fit_transform(df['tweet_lemmatized'])

feature_names_lemmatized = vectorizer_lemmatized.get_feature_names()
pd.DataFrame(bag_of_words_lemmatized.toarray(), columns = feature_names_lemmatized)
```

```python
from sklearn import model_selection, preprocessing, linear_model
from sklearn.metrics import *
```

```python
data = open('corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))


corpus_df = pd.DataFrame()
corpus_df['text'] = texts
corpus_df['label'] = labels
corpus_df.head(3)
```

```python
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(corpus_df['text'], corpus_df['label'])

encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)
```

```python
count_vectorizer = CountVectorizer(max_df=0.9) 
count_vectorizer.fit(corpus_df['text'])

xtrain_count =  count_vectorizer.transform(train_x)
xvalid_count =  count_vectorizer.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions = classifier.predict(xvalid_count)

acc_score = accuracy_score(valid_y, predictions)
acc_score 
```

```python
print(classification_report(predictions,valid_y))
```

```python
results = pd.DataFrame() 

res1 = {'Type':'CountVectorizer', 'Score':acc_score}

results = results.append(res1, ignore_index=True)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.9)
tfidf_vectorizer.fit(corpus_df['text'])

xtrain_tfidf =  tfidf_vectorizer.transform(train_x)
xvalid_tfidf =  tfidf_vectorizer.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_tfidf, train_y)
predictions = classifier.predict(xvalid_tfidf)

acc_score = accuracy_score(valid_y, predictions)
acc_score 
```

```python
print(classification_report(predictions,valid_y))
```

```python
res2 = {'Type':'TfidfVectorizer', 'Score':acc_score}

results = results.append(res2, ignore_index=True)
```
